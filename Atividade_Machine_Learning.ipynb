```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.ensemble import RandomForestClassifier
```

```python
base = pd.read_csv("DADOS/soybean.csv")
base.head()
```

```python
previsores = base.iloc[:,0:35].values
classe = base.iloc[:,35].values
previsores
```

```python
classe
```

```python
labelencoder = LabelEncoder()
for x in range(35):
    previsores[:, x] = labelencoder.fit_transform(previsores[:, x])
```

```python
x_treino, x_teste, y_treino, y_teste = train_test_split(previsores,
                                                       classe,
                                                       test_size = 0.3,
                                                       random_state = 0)
x_teste
```

```python
x_treino
```

```python
floresta = RandomForestClassifier(n_estimators = 100)
floresta.fit(x_treino, y_treino)
```

```python
floresta.estimators_
```

```python
previsoes = floresta.predict(x_teste)
previsoes
```

```python
confusao = confusion_matrix(y_teste, previsoes)
confusao
```

```python
taxa_acerto = accuracy_score(y_teste, previsoes)
taxa_acerto
```

```python
taxa_erro = 1 - taxa_acerto
taxa_erro
```
