```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
import matplotlib.pyplot as plt
```

```python
# 1. Criando um dataset simples
dados = {
    'idade': [25, 40, 35, 28, 50, 45, 23, 60, 30, 33],
    'renda_mensal': [3000, 8000, 5000, 3200, 10000, 9500, 2500, 12000, 4000, 4500],
    'divida': [0, 2000, 1000, 0, 4000, 3500, 0, 6000, 500, 700],
    'score_credito': [600, 750, 720, 610, 800, 770, 580, 850, 690, 700],
    'aprovado': [0, 1, 1, 0, 1, 1, 0, 1, 1, 1]  # variável alvo
}

df = pd.DataFrame(dados)

# 2. Separando os dados
X = df[['idade', 'renda_mensal', 'divida', 'score_credito']]
y = df['aprovado']

# 3. Dividindo em treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 4. Treinando o modelo
modelo = RandomForestClassifier()
modelo.fit(X_train, y_train)

# 5. Fazendo previsões
y_pred = modelo.predict(X_test)

# 6. Avaliando
print("Acurácia:", accuracy_score(y_test, y_pred))
print("Relatório de Classificação:\n", classification_report(y_test, y_pred))

# 7. Visualizando importância das features
importancias = modelo.feature_importances_
plt.bar(X.columns, importancias)
plt.title('Importância das Features')
plt.ylabel('Importância')
plt.show()
```
